# 🧪 Módulo 00 — Creación Básica de Agentes

> **Objetivo:** Aprender las diferentes formas de crear un agente con Microsoft Agent Framework para Python.

## ¿Qué vas a aprender?

- Cómo construir un `Agent` con un `ChatClient` de Azure OpenAI
- Cómo personalizar nombre, descripción e instrucciones (system prompt)
- Cómo ejecutar el agente con `await agent.run("...")` y obtener una respuesta

## Conceptos clave

| Concepto | API |
|----------|-----|
| Cliente Azure OpenAI | `OpenAIChatClient(api_key=..., azure_endpoint=..., model=...)` |
| Agente | `Agent(client=..., instructions=..., name=..., description=...)` |
| Ejecución | `response = await agent.run("prompt"); response.text` |


## ⚙️ Setup inicial

Esta celda carga la configuración de Azure OpenAI desde `appsettings.Development.json`
(o variables de entorno) y crea el cliente que reutilizaremos durante todo el notebook.

> 💡 Solo necesitas ejecutar esta celda **una vez** al abrir el notebook.


```python
# Aseguramos que la raíz del proyecto esté en sys.path para importar `helpers`
import sys
from pathlib import Path

_ROOT = Path.cwd()
# Si abriste el notebook desde la carpeta notebooks/, sube un nivel
if _ROOT.name == "notebooks":
    _ROOT = _ROOT.parent
if str(_ROOT) not in sys.path:
    sys.path.insert(0, str(_ROOT))

from agent_framework import Agent
from helpers.config import create_chat_client
print("✅ Helpers cargados. Endpoint:", create_chat_client().__class__.__name__)
```

## 1️⃣ Crear un agente con opciones personalizadas

Las **instrucciones** definen el "system prompt" — el comportamiento y personalidad del agente.
El **nombre** y la **descripción** son útiles cuando coordinas varios agentes (ver Módulos 12-13).


```python
agent = Agent(
    client=create_chat_client(),
    instructions="Eres un asistente de taller. Siempre responde de forma concisa y educativa.",
    name="AsistenteWorkshop",
    description="Agente de demostración para el taller de Agent Framework",
)

print(f"✅ Agente '{agent.name}' creado")
print(f"   Descripción: {agent.description}")
```

## 2️⃣ Enviar un mensaje y recibir una respuesta

Llamar a `await agent.run("...")` devuelve un objeto `AgentResponse` con la propiedad `.text`.

> Sin `session=...` cada llamada es **independiente** (el agente no recuerda turnos anteriores).
> Para conversaciones multi-turno usa sesiones — lo veremos en el **Módulo 06**.


```python
agent = Agent(
    client=create_chat_client(),
    instructions="Eres un asistente útil. Siempre responde en una oración corta.",
)

response = await agent.run("¿Cuánto es 2 + 2?")
print(f"✅ Respuesta del agente: {response.text}")
```

## 3️⃣ Crear el `ChatClient` por separado

Este patrón es útil cuando **varios agentes comparten el mismo cliente** (ahorra conexiones y reutiliza configuración).


```python
from agent_framework import Agent

# Crear el client una sola vez
client = create_chat_client()

# Usarlo en uno o varios agentes
agent = Agent(
    client=client,
    name="AgenteExplicito",
    instructions="Eres un asistente conciso. Responde en una sola línea.",
)

response = await agent.run("Dime el nombre de un planeta.")
print(f"✅ Agente '{agent.name}' usando ChatClient compartido")
print(f"   Respuesta: {response.text}")
```

## 🎯 Resumen

| Capacidad | Cómo se hace |
|-----------|--------------|
| Crear agente con instrucciones | `Agent(client=..., instructions="...", name="...")` |
| Obtener respuesta simple | `response = await agent.run("..."); response.text` |
| Reutilizar `ChatClient` | Crearlo una vez y pasárselo a varios `Agent(...)` |
